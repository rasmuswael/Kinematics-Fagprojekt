```python
df = pd.read_excel("./survey-data.xlsx")
```

```python
CMUdf = df.loc[df['Model'] == "Original CMU"]
HMMdf = df.loc[df['Model'] == "HMM"]
Normaldf = df.loc[df['Model'] == "Normal"]
```

```python
plt.hist(Normaldf["Score"])
np.mean(Normaldf['Score'])
```

```python
HMMdf['Model'] = HMMdf['Model'] +' '+ HMMdf['Label']
CMUdf['Model'] = CMUdf['Model'] +' '+ CMUdf['Label']
Normaldf['Model'] = Normaldf['Model'] +' '+ Normaldf['Label']



```

```python
sns.set_style("whitegrid")

fig, ax =plt.subplots(1,3, sharey = True)
fig.set_size_inches(16, 4)

sns.histplot(data=CMUdf, x="Score", hue="Model", discrete=True, multiple = 'dodge', shrink = 0.8, ax = ax[0], palette = 'mako')
ax[0].set_title("Motion Capture Data (control)",fontweight = 'bold', fontsize=15)
ax[0].set_xlabel("")
ax[0].tick_params(axis='x', labelsize= 14)

sns.histplot(data=Normaldf, x="Score", hue="Model", discrete=True, multiple = 'dodge', shrink = 0.8, ax = ax[1], palette = 'mako')

ax[1].set_title("Multivariate Gaussian Model",fontweight = 'bold', fontsize=15)
ax[1].set_xlabel("")
ax[1].tick_params(axis='x', labelsize= 14)

sns.histplot(data=HMMdf, x="Score", hue="Model", discrete=True, multiple = 'dodge', shrink = 0.8, ax = ax[2], palette = 'mako')

ax[2].set_title("Gaussian HMM Model",fontweight = 'bold', fontsize=15)
ax[2].set_xlabel("")
ax[2].tick_params(axis='x', labelsize= 14)


fig.text(0.5, -0.02, 'Scores', ha='center', fontsize = 14)
ax[0].set_ylabel("Count", fontsize=14)

fig.tight_layout()
fig.show()

```
